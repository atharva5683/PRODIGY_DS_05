```python
plt.figure(figsize=(10,3))
sns.countplot(x = 'Day_of_week', data=df)
plt.show()
```

```python
plt.figure(figsize=(100, 90))
sns.countplot(x = "Area_accident_occured", data=df)
plt.show()
```

```python
plt.figure(figsize=(46,17))
sns.countplot(x='Type_of_vehicle', data=df)
plt.show()
```

```python
plt.figure(figsize=(20,8))
ax = sns.countplot(x='Time', data=df, order=sorted(df['Time'].unique()))
ax.set_xticks(ax.get_xticks()[::10])
plt.xticks(rotation=90)
plt.show()
```

```python
average_severity_by_weather = df.groupby('Weather_conditions')['Accident_severity'].value_counts(normalize=True).unstack().mean(axis=1)
print(average_severity_by_weather)
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
